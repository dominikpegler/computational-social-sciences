## NewscatcherApi (new)

```python
from newscatcherapi import NewsCatcherApiClient

nc_api_key = open("newscatcherapi_key.txt", "r").readlines()[0]

newscatcherapi = NewsCatcherApiClient(x_api_key=nc_api_key)
```

latest

```python
top_headlines = newscatcherapi.get_latest_headlines(
    lang="de", countries="at", topic="politics"
)

for a in top_headlines["articles"]:
    print(a["published_date"], "|", a["topic"], "|", a["clean_url"], "|", a["title"])
```

search all articles

Only the last 30 days with free API key

```python
all_articles = newscatcherapi.get_search(
    q="Teichtmeister", lang="de", countries="AT", from_="2022/12/22", to_="2023/1/19"
)

for a in all_articles["articles"]:
    print(a["published_date"], "|", a["topic"], "|", a["clean_url"], "|", a["title"])
```

## Feedparser

for RSS feeds

```python
# !pip install feedparser feedsearch

from feedsearch import search
import feedparser
import time


feeds = search("https://www.derstandard.at/rss")
urls = [f.url for f in feeds]

print(urls)

if len(urls) > 0:
    news_feed = feedparser.parse(urls[0])
    entries = news_feed["entries"]
    count = 0
    for entry in entries:
        count = count + 1
        print(
            str(count)
            + ". "
            + entry["title"]
            + "\n\t\t"
            + entry["published"]
            + "\n\t\t"
            + entry["link"]
            + "\n\n"
        )
        time.sleep(0.33)
```

## newspaper3k

```python
# !pip install newspaper3k

import newspaper
import json

url = "https://www.derstandard.at"

article = newspaper.Article(url=url, language="de")
article.download()
article.parse()

article = {
    "title": str(article.title),
    "text": str(article.text),
    "authors": article.authors,
    "published_date": str(article.publish_date),
    "top_image": str(article.top_image),
    "videos": article.movies,
    "keywords": article.keywords,
    "summary": str(article.summary),
}


print(
    article["title"]
    + "\n\t\t"
    + article["published_date"]
    + "\n\n"
    + "\n"
    + article["text"]
    + "\n\n"
)
```
